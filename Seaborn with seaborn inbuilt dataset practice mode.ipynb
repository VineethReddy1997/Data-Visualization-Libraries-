# Dist plots

Dist plots helps us to checks the distribution of the columns feature

```python
sns.distplot(df['tip'])  # it will be create histograms 
```

```python
 sns.distplot(df['tip'],kde=False,bins=10)  # kde is kernel density estimation
```

# Categorical Plots

Count plot will be showing the single categorical feagtures as like discrete values or finite values  finding purpose 

```python
#count plot
sns.countplot(x= 'sex',data = df)  #vertical  count  plot by default
```

```python
#count plot    Horizontally (y-axis)
sns.countplot(y= 'sex',data = df)
```

```python
#count plot
sns.countplot(x= 'smoker',data = df)    #smokers features
```

```python
#count plot
sns.countplot(x= 'size',data = df)  #size count plots
```

```python
#count plot
sns.countplot(y= 'day',data = df)  #day feature count plots
```

Bar plots have check  to take two features relations in data

# Bar plot

```python
#bar plot    This plot will be use in categorical featues
sns.barplot(x='sex',y = 'total_bill',data=df)
```

```python
sns.barplot(x='smoker',y = 'total_bill',data=df)
```

```python
sns.barplot(x='day',y = 'total_bill',data=df)
```

```python
sns.barplot(y='size',x = 'day',hue = "smoker",data=df)
```

# Box Plot 

A box and whisker plot (sometimes called a boxplot ) it is a graph  that presents information from a five number summary

```python
sns.boxplot(x = 'sex', y = 'total_bill',data = df)
```

```python
sns.boxplot(x = 'day', y = 'total_bill',data = df)
```

```python
sns.boxplot(y = 'day',x = 'total_bill',data = df)
```

```python
sns.boxplot(y = 'day',x = 'total_bill',data = df,palette='twilight_shifted_r')    #with color viewing
```

```python
sns.boxplot(data=df,orient="h") #h msens horzontally viewing the data  points 
```

```python
# categorize my data based on  some other  categories

sns.boxplot(x = 'total_bill', y='day',hue = 'smoker',data=df)
```

# violin plot

```python
sns.violinplot(x = 'total_bill',y = 'day',hue ='smoker',data = df)
```
